```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import sem
```

```python
pd_syscall = pd.read_csv("data/COMP410_2.csv")
pd_syscall
```

```python
writehello_mean = pd_syscall.loc[pd_syscall["syscall"] == "writeHello", "LTCode"].sum() / pd_syscall.loc[pd_syscall["syscall"] == "writeHello", "LTCode"].count()
writehello_std = pd_syscall.loc[pd_syscall["syscall"] == "writeHello", "LTCode"].std()

writelorem1_mean = pd_syscall.loc[pd_syscall["syscall"] == "writeLorem1", "LTCode"].sum() / pd_syscall.loc[pd_syscall["syscall"] == "writeLorem1", "LTCode"].count()
writelorem1_std = pd_syscall.loc[pd_syscall["syscall"] == "writeLorem1", "LTCode"].std()

writelorem10_mean = pd_syscall.loc[pd_syscall["syscall"] == "writeLorem10", "LTCode"].sum() / pd_syscall.loc[pd_syscall["syscall"] == "writeLorem10", "LTCode"].count()
writelorem10_std = pd_syscall.loc[pd_syscall["syscall"] == "writeLorem10", "LTCode"].std()

open_mean = pd_syscall.loc[pd_syscall["syscall"] == "openPass", "LTCode"].sum() / pd_syscall.loc[pd_syscall["syscall"] == "openPass", "LTCode"].count()
open_std = pd_syscall.loc[pd_syscall["syscall"] == "openPass", "LTCode"].std()

fork0_mean = pd_syscall.loc[pd_syscall["syscall"] == "fork0", "LTCode"].sum() / pd_syscall.loc[pd_syscall["syscall"] == "fork0", "LTCode"].count()
fork0_std = pd_syscall.loc[pd_syscall["syscall"] == "fork0", "LTCode"].std()

fork1_mean = pd_syscall.loc[pd_syscall["syscall"] == "fork1", "LTCode"].sum() / pd_syscall.loc[pd_syscall["syscall"] == "fork1", "LTCode"].count()
fork1_std = pd_syscall.loc[pd_syscall["syscall"] == "fork1", "LTCode"].std()

fork2_mean = pd_syscall.loc[pd_syscall["syscall"] == "fork2", "LTCode"].sum() / pd_syscall.loc[pd_syscall["syscall"] == "fork2", "LTCode"].count()
fork2_std = pd_syscall.loc[pd_syscall["syscall"] == "fork2", "LTCode"].std()

writehello_ste = writehello_std / np.sqrt(20)
writelorem1_ste = writelorem1_std / np.sqrt(20)
writelorem10_ste = writelorem10_std / np.sqrt(20)
open_ste = open_std / np.sqrt(20)
fork0_ste = fork0_std / np.sqrt(20)
fork1_ste = fork1_std / np.sqrt(20)
fork2_ste = fork2_std / np.sqrt(20)

print("writeHello: mean = " + str(writehello_mean) + ", std = " + str(writehello_std) + ", ste = " + str(writehello_ste))
print("writeLorem1: mean = " + str(writelorem1_mean) + ", std = " + str(writelorem1_std) + ", ste = " + str(writelorem1_ste))
print("writeLorem10: mean = " + str(writelorem10_mean) + ", std = " + str(writelorem10_std) + ", ste = " + str(writelorem10_ste))
print("open: mean = " + str(open_mean) + ", std = " + str(open_std) + ", ste = " + str(open_ste))
print("fork 0: mean = " + str(fork0_mean) + ", std = " + str(fork0_std) + ", ste = " + str(fork0_ste))
print("fork 1: mean = " + str(fork1_mean) + ", std = " + str(fork1_std) + ", ste = " + str(fork1_ste))
print("fork 2: mean = " + str(fork2_mean) + ", std = " + str(fork2_std) + ", ste = " + str(fork2_ste))

```

```python
iteration = ["Open", "Write - 12 bytes", "Write - 815 bytes", "Write - 8.2kb"]
lt_mean = [open_mean, writehello_mean, writelorem1_mean, writelorem10_mean]
lt_ste = [open_ste, writehello_ste, writelorem1_ste, writelorem10_ste]

fig, ax = plt.subplots()
ax.bar(x=np.arange(len(iteration)), #x-coordinates of bars
       height=lt_mean, #height of bars
       yerr=lt_ste, #error bar width
       capsize=20) #length of error bar caps

x_axis2 = np.arange(len(iteration))

plt.xticks(x_axis2, iteration, rotation = 45)
plt.title("Average processing time for syscalls")
plt.xlabel("syscall")
plt.ylabel("processing time")
plt.figure(figsize = (20, 5))

plt.show()
```

```python
iteration = ["No fork", "1 Fork", "2 Forks"]
lt_mean = [fork0_mean, fork1_mean, fork2_mean]
lt_ste = [fork0_ste, fork1_ste, fork2_ste]

fig = plt.figure(figsize = (5, 10))
ax = fig.add_subplot(111)
ax.bar(x=np.arange(len(iteration)), #x-coordinates of bars
       height=lt_mean, #height of bars
       yerr=lt_ste, #error bar width
       capsize=20) #length of error bar caps

x_axis2 = np.arange(len(iteration))


plt.xticks(x_axis2, iteration, rotation = 45)
plt.yticks([0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9])
plt.ylim(0.3, 0.9)
plt.title("Average processing time for fork() syscalls")
plt.xlabel("amount of forks")
plt.ylabel("processing time")

plt.show()
```

```python
# Latency Means
lat_4byte_mean = pd_socket.loc[pd_socket["size"] == "4byte", "LT_Latency"].sum() / pd_socket.loc[pd_socket["size"] == "4byte", "LT_Latency"].count()
lat_16byte_mean = pd_socket.loc[pd_socket["size"] == "16byte", "LT_Latency"].sum() / pd_socket.loc[pd_socket["size"] == "16byte", "LT_Latency"].count()
lat_64byte_mean = pd_socket.loc[pd_socket["size"] == "64byte", "LT_Latency"].sum() / pd_socket.loc[pd_socket["size"] == "64byte", "LT_Latency"].count()
lat_256byte_mean = pd_socket.loc[pd_socket["size"] == "256byte", "LT_Latency"].sum() / pd_socket.loc[pd_socket["size"] == "256byte", "LT_Latency"].count()
lat_1024byte_mean = pd_socket.loc[pd_socket["size"] == "1024byte", "LT_Latency"].sum() / pd_socket.loc[pd_socket["size"] == "1024byte", "LT_Latency"].count()
lat_4k_mean = pd_socket.loc[pd_socket["size"] == "4k", "LT_Latency"].sum() / pd_socket.loc[pd_socket["size"] == "4k", "LT_Latency"].count()
lat_16k_mean = pd_socket.loc[pd_socket["size"] == "16k", "LT_Latency"].sum() / pd_socket.loc[pd_socket["size"] == "16k", "LT_Latency"].count()
lat_64k_mean = pd_socket.loc[pd_socket["size"] == "64k", "LT_Latency"].sum() / pd_socket.loc[pd_socket["size"] == "64k", "LT_Latency"].count()

# Throughput Means
thp_4byte_mean = pd_socket.loc[pd_socket["size"] == "4byte", "LT_Throughput"].sum() / pd_socket.loc[pd_socket["size"] == "4byte", "LT_Throughput"].count()
thp_16byte_mean = pd_socket.loc[pd_socket["size"] == "16byte", "LT_Throughput"].sum() / pd_socket.loc[pd_socket["size"] == "16byte", "LT_Throughput"].count()
thp_64byte_mean = pd_socket.loc[pd_socket["size"] == "64byte", "LT_Throughput"].sum() / pd_socket.loc[pd_socket["size"] == "64byte", "LT_Throughput"].count()
thp_256byte_mean = pd_socket.loc[pd_socket["size"] == "256byte", "LT_Throughput"].sum() / pd_socket.loc[pd_socket["size"] == "256byte", "LT_Throughput"].count()
thp_1024byte_mean = pd_socket.loc[pd_socket["size"] == "1024byte", "LT_Throughput"].sum() / pd_socket.loc[pd_socket["size"] == "1024byte", "LT_Throughput"].count()
thp_4k_mean = pd_socket.loc[pd_socket["size"] == "4k", "LT_Throughput"].sum() / pd_socket.loc[pd_socket["size"] == "4k", "LT_Throughput"].count()
thp_16k_mean = pd_socket.loc[pd_socket["size"] == "16k", "LT_Throughput"].sum() / pd_socket.loc[pd_socket["size"] == "16k", "LT_Throughput"].count()
thp_64k_mean = pd_socket.loc[pd_socket["size"] == "64k", "LT_Throughput"].sum() / pd_socket.loc[pd_socket["size"] == "64k", "LT_Throughput"].count()

print("4byte Latency: " + str(lat_4byte_mean) + ", 4byte Throughput: " + str(thp_4byte_mean))
print("16byte Latency: " + str(lat_16byte_mean) + ", 4byte Throughput: " + str(thp_16byte_mean))
print("64byte Latency: " + str(lat_64byte_mean) + ", 4byte Throughput: " + str(thp_64byte_mean))
print("256byte Latency: " + str(lat_256byte_mean) + ", 4byte Throughput: " + str(thp_256byte_mean))
print("1024byte Latency: " + str(lat_1024byte_mean) + ", 4byte Throughput: " + str(thp_1024byte_mean))
print("4k Latency: " + str(lat_4k_mean) + ", 4byte Throughput: " + str(thp_4k_mean))
print("16k Latency: " + str(lat_16k_mean) + ", 4byte Throughput: " + str(thp_16k_mean))
print("64k Latency: " + str(lat_64k_mean) + ", 4byte Throughput: " + str(thp_64k_mean))
```

```python
lat_x = ["4-byte", "16-byte", "64-byte", "256-byte", "1024-byte", "4kB", "16kB", "64kB"]
lat_y = [lat_4byte_mean, lat_16byte_mean, lat_64byte_mean, lat_256byte_mean, lat_1024byte_mean, lat_4k_mean, lat_16k_mean, lat_64k_mean]


plt.plot(lat_x, lat_y)
plt.title("Socketpair Latency between Parent & Child Processes")
plt.xlabel("Message size")
plt.xticks(rotation = 45)
plt.ylabel("Latency (ns)")
plt.show()
```
