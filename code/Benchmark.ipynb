```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
pd = pd.read_csv("data/COMP410_1.csv")
pd
```

```python
pd_sub = pd[["Iterations", "LTCode", "PCCode"]]
pd_sub
```

```python
LT100avg = pd_sub.loc[pd_sub["Iterations"] == "100", "LTCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "100", "LTCode"].count()
LT1Kavg = pd_sub.loc[pd_sub["Iterations"] == "1K", "LTCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "1K", "LTCode"].count()
LT10Kavg = pd_sub.loc[pd_sub["Iterations"] == "10K", "LTCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "10K", "LTCode"].count()
LT100Kavg = pd_sub.loc[pd_sub["Iterations"] == "100K", "LTCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "100K", "LTCode"].count()
LT1Mavg = pd_sub.loc[pd_sub["Iterations"] == "1M", "LTCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "1M", "LTCode"].count()
LT10Mavg = pd_sub.loc[pd_sub["Iterations"] == "10M", "LTCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "10M", "LTCode"].count()
LT100Mavg = pd_sub.loc[pd_sub["Iterations"] == "100M", "LTCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "100M", "LTCode"].count()
LT1Bavg = pd_sub.loc[pd_sub["Iterations"] == "1B", "LTCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "1B", "LTCode"].count()


PC100avg = pd_sub.loc[pd_sub["Iterations"] == "100", "PCCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "100", "PCCode"].count()
PC1Kavg = pd_sub.loc[pd_sub["Iterations"] == "1K", "PCCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "1K", "PCCode"].count()
PC10Kavg = pd_sub.loc[pd_sub["Iterations"] == "10K", "PCCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "10K", "PCCode"].count()
PC100Kavg = pd_sub.loc[pd_sub["Iterations"] == "100K", "PCCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "100K", "PCCode"].count()
PC1Mavg = pd_sub.loc[pd_sub["Iterations"] == "1M", "PCCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "1M", "PCCode"].count()
PC10Mavg = pd_sub.loc[pd_sub["Iterations"] == "10M", "PCCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "10M", "PCCode"].count()
PC100Mavg = pd_sub.loc[pd_sub["Iterations"] == "100M", "PCCode"].sum() / pd_sub.loc[pd_sub["Iterations"] =="100M", "PCCode"].count()
PC1Bavg = pd_sub.loc[pd_sub["Iterations"] == "1B", "PCCode"].sum() / pd_sub.loc[pd_sub["Iterations"] == "1B", "PCCode"].count()
PC1Bavg
```

```python
iteration = ["$10^2$", "$10^3$", "$10^4$", "$10^5$", "$10^6$", "$10^7$", "$10^8$", "$10^9$"]
lt_mean = [LT100avg, LT1Kavg, LT10Kavg, LT100Kavg, LT1Mavg, LT10Mavg, LT100Mavg, LT1Bavg]
pc_mean = [PC100avg, PC1Kavg, PC10Kavg, PC100Kavg, PC1Mavg, PC10Mavg, PC100Mavg, PC1Bavg]

x_axis = np.arange(len(iteration)) 
  
plt.bar(x_axis - 0.2, lt_mean, 0.4, label = 'Laptop') 
plt.bar(x_axis + 0.2, pc_mean, 0.4, label = 'PC') 
  
plt.xticks(x_axis, iteration) 
plt.xlabel("iterations") 
plt.ylabel("processing time") 
plt.title("Average process time per amount of iterations (Built-In)") 
plt.legend() 
plt.show() 
```
